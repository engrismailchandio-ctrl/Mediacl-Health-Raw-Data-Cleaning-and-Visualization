```python
# 1. Get the proportions (ignoring NaNs)
probs = df['smoker'].value_counts(normalize=True)

# 2. Identify where the missing values are
missing_mask = df['smoker'].isna()
n_missing = missing_mask.sum()

# 3. Generate 50/50 "yes" and "no" values for the 20% gaps
fill_values = np.random.choice(probs.index, size=n_missing, p=probs.values)

# 4. Fill only the empty slots
df.loc[missing_mask, 'smoker'] = fill_values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtering for only the 'Adult' group (optional if you want only adults)
adult_data = df[df['age_group'] == 'adult']

# Creating a faceted count plot
sns.catplot(
    data=adult_data, 
    x='gender', 
    hue='diagnosis',      # Color bars by disease status
    kind='count',        # Use 'count' to show patient numbers
    palette='viridis'    # Choose a clear color palette
)

plt.title('Diabetes Status by Gender (Adult Group)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset=df

# 1. Filter for the 'Adult' age group
adults = dataset[dataset['age_group'] == 'adult']

# 2. Visualize Diagnosis status by Gender
sns.catplot(
    data=adults, 
    x='gender', 
    hue='diagnosis',  # Assuming your column is named 'Diagnosis'
    kind='count', 
    palette='magma',
    height=5, 
    aspect=1.2
)

plt.title('Disease Diagnosis by Gender (Adult Group)')
plt.ylabel('Number of Patients')
plt.show()
```
